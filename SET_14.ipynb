```python
#import Module
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data From URL
!wget https://raw.githubusercontent.com/Abdullahfoysal/pp11-Project-AnalyzeAndVisualize/main/student_data.csv
#Read Data
df = pd.read_csv("student_data.csv")
df.describe()
```

```python
# 1. Count the number of students who have an HSC GPA greater than 3.5.
hsc_gpa_list = df[df["hsc_gpa"] > 3.5]
print(hsc_gpa_list)
```

```python
#2.we can't find chemical_engineering_department.so we calculate students from "Engineering" who have a CGPA above 3.0
engineering_students = df[(df["department"] == "Engineering") & (df["cgpa"] > 3.0)]
print(engineering_students)
```

```python
# 3. Calculate the average attendance for all students who have studied abroad.
# Replace multiple values
df['abroad_study'] = df['abroad_study'].replace({'true': 1})
print(df)
abroad_students = df[df['abroad_study'] == 1]
average_attendance_abroad = abroad_students['attendance'].mean()
print(f"The average attendance for all students who have studied abroad is {average_attendance_abroad}.")
```

```python
#Generate a line chart comparing the number of students from each department over the years.
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and contains 'year' and 'department' columns

# Group data by year and department, count students, and unstack
department_counts = df.groupby(['passing_year', 'department']).size().unstack().fillna(0)

# Create the line chart
department_counts.plot(kind='line', figsize=(10, 6))  # Adjust figsize as needed

# Customize the chart
plt.title('Number of Students from Each Department Over the Passing_Years')
plt.xlabel('passing_year')
plt.ylabel('Number of Students')
plt.legend(title='Department')

# Display the chart
plt.show()
```
