```python
#Import Module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Dataset from URL
!wget https://raw.githubusercontent.com/Abdullahfoysal/pp11-Project-AnalyzeAndVisualize/refs/heads/main/student_data.csv
```

```python
#read dataset
df=pd.read_csv('student_data.csv')
df
```

```python
#1.List all students with CGPA above
cgpa_list=df[df["cgpa"]>3.5]
print(cgpa_list)
```

```python
#2.Find the number of students who graduated after 2020
graduation_list=df[df['graduation_date']>2020]
print(graduation_list)

```

```python
#3. Calculate the average attendance of students in the "Engineering"department.
engineering_student=df[df["department"]=="Engineering"]
print(engineering_student)
average_attendance=engineering_student["attendance"].mean()
print(average_attendance)
```

```python
#4.Repalce missing values in "study_Abroad_country" with "Not Available"
df.filled=df.fillna({"study_abroad_country":"Not Available"})
print(df.filled)
df.filled
```

```python
#5.Generate a line plot to show the trend in CGPA over passing years.
import matplotlib.pyplot as plt
# df_under=df.sample(50)
passing_year=df["passing_year"]
cgpa=df["cgpa"]
plt.plot(passing_year,cgpa,marker="o",linestyle="--",color="g",label="Point")
plt.title("The trend in CGPA over passing years.")
plt.xlabel("passing_year")
plt.ylabel("cgpa")
plt.show()
```

```python
#5.Generate a line plot to show the trend in CGPA over passing years.
import matplotlib.pyplot as plt
# df_under=df.sample(50)
passing_year_list=df[df["passing_year"]>2013]
print(passing_year_list)
cgpa=df["cgpa"]
plt.plot(passing_year_list,cgpa,marker="o",linestyle="--",color="g",label="Point")
plt.title("The trend in CGPA over passing years.")
plt.xlabel("passing_year_list")
plt.ylabel("cgpa")
plt.show()
```
